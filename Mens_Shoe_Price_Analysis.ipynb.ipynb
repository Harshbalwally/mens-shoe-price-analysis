```python
# Download the dataset using kagglehub
# This will download a zip file to a local cache directory
path = kagglehub.dataset_download("sureshmecad/mens-shoe-prices")

# The path variable points to the downloaded directory.
# We need to find the CSV file within this directory.
import os

csv_file_path = ""
# List files in the downloaded directory and find the first CSV file
for filename in os.listdir(path):
    if filename.endswith('.csv'):
        csv_file_path = os.path.join(path, filename)
        break

if csv_file_path:
    print(f"CSV file found: {csv_file_path}")
    # Load the dataset into a pandas DataFrame
    df = pd.read_csv(csv_file_path)
    print("Dataset loaded successfully!")
    display(df.head()) # Use display for better output in Colab
else:
    print("Error: No CSV file found in the downloaded dataset directory.")
```

```python
# Get a summary of the dataframe's structure and data types
print("Dataframe Info:")
df.info()

# Check for missing values
print("\nMissing Values Per Column:")
print(df.isnull().sum())

# --- Data Cleaning Steps ---

# 1. The 'prices_amountmin' column is an object (text). We need to convert it to a number.
# It contains currency symbols, commas, and might have ranges (e.g., "$50 - $60").
# For this analysis, we'll take the first number we find and convert it.
def clean_price(price):
    if pd.isnull(price):
        return None
    # Remove currency symbols, commas, and get the first part of a price range
    price = str(price).replace('$', '').replace(',', '').strip()
    # Handle ranges by taking the lower end of the range
    if '-' in price:
        price = price.split('-')[0].strip()
    try:
        return float(price)
    except ValueError:
        return None # Return None if conversion fails

df['Cleaned_Price'] = df['prices_amountmin'].apply(clean_price)

# 2. Drop rows where the price could not be cleaned
df.dropna(subset=['Cleaned_Price'], inplace=True)

# 3. Clean the 'brand' column by removing leading/trailing spaces
df['brand'] = df['brand'].str.strip()


print("\nData cleaning complete. 'Cleaned_Price' column added.")
display(df[['brand', 'prices_amountmin', 'Cleaned_Price']].head())
```
